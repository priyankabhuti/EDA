1.**Importing Libraries** 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. **Reading CSV**

```python
path = "/content/drive/MyDrive/zomato.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.shape
```

```python
#Dropping the unwanted columns:
df = df.drop(['url','address','phone','dish_liked','menu_item','reviews_list'], axis=1)
df.head()
```

```python
df.info()
```

```python
df.rename(columns= {'approx_cost(for two people)':'cost_2people' , 'listed_in(type)':'type'} , inplace = True)
df.head()
```

3. **Data Cleaning**

```python
#We can see 5 columns have null values, so cleaning them one by one.
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.shape
```

3.1 Cleaning Rate Column

```python
df['rate'].unique()
```

```python
#replacing and cleaning rate column
def cleanrate(value):
    if (value=='NEW' or value=='-'):
     return np.nan
    else:
      value = str(value).split('/')
      value = value[0]
      return float(value)

df['rate'] = df['rate'].apply(cleanrate)
df['rate'].unique()
```

```python
#filling null values
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

3.2 Cleaning Location Column

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
#location and listed_in(city) has similar values dropping one of them.
df = df.drop(['listed_in(city)'], axis=1)
df.head()
```

```python
location = df['location'].value_counts(ascending  = False)
location
```

```python
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
#Combining location values less than 300 as Others
def handle_location(value):
    if(value in location_lessthan300):
        return 'Others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

3.3 Cleaning rest_type Column

```python
df['rest_type'].unique()
```

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
#Since these are very less in quantity making all values less than 1000 as 'Others'
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'Others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

3.4 Cleaning Cuisines Column

```python
df['cuisines'].unique()
```

```python
cuisines = df['cuisines'].value_counts(ascending=False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

```python
#Since these are very less in quantity making all values less than 100 as 'Others'
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'Others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

3.5 Cleaning the cost_2people Cloumn

```python
df['cost_2people'].unique()
```

```python
#To remove ',' from the column:
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['cost_2people'] = df['cost_2people'].apply(handlecomma)
df['cost_2people'].unique()
        
```
